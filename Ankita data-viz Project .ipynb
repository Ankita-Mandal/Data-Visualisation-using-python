### Finding correlation between scores in all three subjects

```python
#Adding a column to store total score to find effect on it too
st['total score'] = st['math score'] + st['reading score'] + st['writing score']
st.head()
```

```python
sns.kdeplot(st['math score'], shade=True)
sns.kdeplot(st['reading score'], shade=True)
sns.kdeplot(st['writing score'], shade=True)
plt.show()
```

Observation: <br> We can notice greater overlapping in reading and writing scores. <br> The overall overlapping in all three scores is also quite high and most students have scored in the similar range of 60-80 marks

```python
st.corr()
```

```python
plt.figure(figsize=(8,3))
plt.title('Covarriance plot of all scores')
sns.heatmap(st.corr())
plt.show()
```

Correlation coefficient shows a measure of the strength of the relationship between two variables. The closer the value is to 1, the stronger is the positive relationship that exists. <br>
Observations: 
- All scores have high correlation with each other, i.e. if a student scores well in one subject, they are likely to score well in all subjects.
- Correlation between reading and writing score is higher than that with math score, hence a student scoring high in either will probably score high in the other. 
- Reading and writing score also have extremely high correlation with the total score. Thus they are a better indicator of total score obtained by student.
- Math score has high correlation with the other scores, with its relation with readinng score is sightly higher

### Analysis of categorical data

```python
st.columns.tolist()
```

```python
g=st['gender'].unique()
r=['group C', 'group D', 'group B', 'group E', 'group A',]
p=['some college', 'associate\'s degree', 'high school', 'some high school', 'bachelor\'s degree  ','master\'s degree']
l=st['lunch'].unique()
t=st['test preparation course'].unique()
```

```python
plt.figure(figsize=(15,5)) 

plt.subplot(1,3,1)
plt.pie(st['gender'].value_counts(), labels= g, autopct= '%0.1f%%')
plt.title("Gender", fontsize=20)

plt.subplot(1,3,2)
plt.pie(st['lunch'].value_counts(), labels= l, autopct= '%0.1f%%')
plt.title("Lunch", fontsize=20)

plt.subplot(1,3,3)
plt.pie(st['test preparation course'].value_counts(), labels= t, autopct= '%0.1f%%')
plt.title("Test preparation course", fontsize=20)
plt.show()

plt.figure(figsize=(15,5))
plt.subplot(1,2,1)
plt.pie(st['race/ethnicity'].value_counts(), labels= r, autopct= '%0.1f%%')
plt.title("Race/Ethnicity", fontsize=20)

plt.subplot(1,2,2)
plt.pie(st['parental level of education'].value_counts(), labels= p, autopct= '%0.1f%%')
plt.title("Parental level of education", fontsize=20)

plt.show()
```

Observation:
- There is a higher proportion of female students(518 or 51.8%) as compared to male students (482 or 48.2%)
- More students have availed the standard lunch (645 students or 64.5%) than the free/reduced lunch (355 students or 35.5%)
- 642 or 64.2% students have not taken a test preparation course while 358 or 35.8% of students have taken such a preparation course
- Most students belong to Group C(319 or 31.9%) followed by Group D (262 or 26.2%), Group B(190 or 19.00%), Group E(140 or 14.0%) and Group A(89 or 8.9%) with the least number of students.
- Most parents have a degree from some college(226 or 22.6%) while least number of them have a master's degree(59 or 5.9%).
